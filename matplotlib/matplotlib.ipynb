# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
data = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv")
data.head()
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
x = data['total_profit'].tolist()
y = data['month_number'].tolist()
plt.plot(y,x)
plt.xlabel('Month number')
plt.ylabel('Profit')
plt.xticks(y)
plt.yticks([100000,200000,300000,400000,500000])
plt.title('Company profit per month')
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
# from turtle import color

# from matplotlib import markers
# from matplotlib.lines import lineStyles


plt.plot(y,x,label='Profit data of last year', color='r',marker='o',linestyle='--')
plt.xticks(y)
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc='lower right')
plt.title('Company profit per month')
plt.show()

```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
facecream = data['facecream'].to_list()
facewash = data['facewash'].to_list()
toothpaste = data['toothpaste'].to_list()
bathingsoap = data['bathingsoap'].to_list()
shampoo = data['shampoo'].to_list()
moisturizer = data['moisturizer'].to_list()
plt.plot(y,facecream,label='Face cream Sales Data',marker='o',color='blue')
plt.plot(y,facewash,label='Face wash Sales Data',marker='o',color='orange')
plt.plot(y,toothpaste,label='Tooth Paste Sales Data',marker='o',color='green')
plt.plot(y,bathingsoap,label='Bathing Soap Sales Data',marker='o',color='r')
plt.plot(y,shampoo,label='Shampoo Sales Data',marker='o',color='purple')
plt.plot(y,moisturizer,label='Móiturizer Sales Data',marker='o',color='brown')
plt.legend(loc='upper left')
plt.xticks(y)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
plt.scatter(y,toothpaste,label='Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title('Tooth paste Sales data')
plt.grid(linewidth=1,linestyle='--')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
plt.bar(y,facecream,width=0.3,label='Face Cream Sales data',align='edge')
plt.bar(y,facewash,width=-0.3,label='Face Wash sales data',align='edge')
plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and Facecream sales data')
plt.grid(True,linewidth=1,linestyle='--')
plt.xticks(y)
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
range=[150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(x,range,label='Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.title('Profit data')
plt.xticks(range)
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
label = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
sale =[data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(),data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
plt.pie(sale,labels=label,autopct='%1.1f%%')
plt.axis('equal')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
